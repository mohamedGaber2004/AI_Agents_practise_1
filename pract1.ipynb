## TypeDict

```python
from typing import TypedDict

class Student(TypedDict):
    name : str 
    age : int
```

```python
Student(name="Ahmed" , age=30),Student(esm = 7, sen = "50")
```

## Union

```python
from typing import Union

def addition (x:Union[int,float]) : 
    return x*2
```

```python
addition(3.7),addition(2)
```

## Optional 

```python
from typing import Optional

def great(name:Optional[str]) : 
    if name is None : 
        print("none")
    else : 
        print(f"hello,{name}")
```

```python
great("ahmed"),great(None)
```
